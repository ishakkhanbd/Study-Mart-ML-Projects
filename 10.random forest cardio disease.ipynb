```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,12]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)

xtrain
```
